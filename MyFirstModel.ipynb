```
import numpy as np
import tensorflow 
from tensorflow import keras

#model to get y=2x-1
model = tensorflow.keras.Sequential(keras.layers.Dense(units=1, input_shape=[1]))

#compile and specify optimizer and loss function
model.compile(optimizer='sgd',loss='mean_squared_error')

#input is xs and output array is ys
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype= float)
ys = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)

#train the model for given number of iterations or epochs
model.fit(xs,ys,epochs=100)

#run the prediction for model to get the nearest answer
print(model.predict([3.0]))
```
